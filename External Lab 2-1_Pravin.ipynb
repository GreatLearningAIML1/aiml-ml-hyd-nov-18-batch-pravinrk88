### Create 500 random samples from normal distribution with mean 1 and  standard deviation 0.5

```python
# Method 1 Plot the 500 random samples from normal distribution with mean 1 and standard deviation 0.5

mu,sigma=1,0.5
s=np.random.normal(mu,sigma,500)
s
```

```python
# Method 2 Plot the 500 random samples from normal distribution with mean 1 and standard deviation 0.5
mu,sigma=1,0.5
s=mu+sigma*np.random.randn(500)
s
```
